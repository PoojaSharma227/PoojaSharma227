```python
# removing Id column
new_data = dataset[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
print(new_data.head())
```

```python
plt.figure(figsize = (10, 7))
new_data.boxplot()
```

```python
#finding the IQR
percentile25 = new_data['SepalWidthCm'].quantile(0.25)
percentile75 = new_data['SepalWidthCm'].quantile(0.75)
print(percentile25)
print(percentile75)
```

```python
iqr = percentile75 - percentile25
iqr
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
print("upper_limit", upper_limit)
print("lower_limit", lower_limit)
```

```python
condition = new_data[new_data['SepalWidthCm'] > upper_limit]
```

```python
condition = new_data[new_data['SepalWidthCm'] < lower_limit]
```

```python
Dataset_cap = new_data.copy()

Dataset_cap['SepalWidthCm'] = np.where(
    Dataset_cap['SepalWidthCm'] > upper_limit,
    upper_limit,
    np.where(
        Dataset_cap['SepalWidthCm'] < lower_limit,
        lower_limit,
        Dataset_cap['SepalWidthCm']
    )
)
```

```python
np.where(condition,True,False)
```
